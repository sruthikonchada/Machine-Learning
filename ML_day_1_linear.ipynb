```python
import pandas as pd
```

```python
file=pd.read_csv("/content/Simple linear regression.csv")
```

```python
file.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(file["SAT"],file["GPA"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(file[["SAT"]],file[["GPA"]])
```

```python
y_pred = model.predict(file[['SAT']])
r_squared = model.score(file[['SAT']], file[['GPA']])
print(f"R-squared score: {r_squared:.4f}")
```

```python
model.predict([[1760]])
```

```python
import matplotlib.pyplot as plt
plt.scatter(file["SAT"], file["GPA"], label='Actual Data')
plt.plot(file["SAT"], y_pred, color='red', label='Regression Line')
plt.legend()

```
